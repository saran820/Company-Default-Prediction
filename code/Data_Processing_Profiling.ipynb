```python
# Loading Datasets
data_1 = pd.DataFrame(arff.loadarff(os.path.join(path, '1year.arff'))[0])
data_1['df_year'] = '1_year'
data_1['class'].replace(to_replace = b'0', value = 0, inplace = True)
data_1['class'].replace(to_replace = b'1', value = 1, inplace = True)

data_2 = pd.DataFrame(arff.loadarff(os.path.join(path, '2year.arff'))[0])
data_2['df_year'] = '2_year'
data_2['class'].replace(to_replace = b'0', value = 0, inplace = True)
data_2['class'].replace(to_replace = b'1', value = 1, inplace = True)

data_3 = pd.DataFrame(arff.loadarff(os.path.join(path, '3year.arff'))[0])
data_3['df_year'] = '3_year'
data_3['class'].replace(to_replace = b'0', value = 0, inplace = True)
data_3['class'].replace(to_replace = b'1', value = 1, inplace = True)

data_4 = pd.DataFrame(arff.loadarff(os.path.join(path, '4year.arff'))[0])
data_4['df_year'] = '4_year'
data_4['class'].replace(to_replace = b'0', value = 0, inplace = True)
data_4['class'].replace(to_replace = b'1', value = 1, inplace = True)

data_5 = pd.DataFrame(arff.loadarff(os.path.join(path, '5year.arff'))[0])
data_5['df_year'] = '5_year'
data_5['class'].replace(to_replace = b'0', value = 0, inplace = True)
data_5['class'].replace(to_replace = b'1', value = 1, inplace = True)

# Combining dataset
data_all = pd.concat([data_1, data_2, data_3, data_4, data_5], axis=0)
```

```python
data_all.head(2)
```

```python
data_all.to_csv(os.path.join(path, 'combined_dataset.csv'))
```

### Data Pre-processing
1.   Removing Attributes with very large missing %
2.   Removing Attributes with large 0 values
3.   Missing Value Imputation using Median


<font color='blue'>*Data: Year1*</font>

---



```python
profile = ProfileReport(data_1, minimal=True)
profile.to_file(os.path.join(path, "report_1.html"))
```

```python
'''Data Year1'''
# Attr37 has 2740 (39.0%) missing 
# Attr21 has 1622 (23.1%) missing values
# Attr6 has 2675 (38.1%) zeros
# Attr59 has 2743 (39.0%) zeros
data_1 = data_1.drop(columns= ['Attr37', 'Attr21', 'Attr6','Attr59'], axis=1)

# Most of the features are skewed hence median is a better imputation strategy
data_1 = data_1.fillna(data_1.median()) 
```

```python
# Any nulls (check)
data_1.columns[data_1.isna().any()]
```

<font color='blue'>*Data: Year2*</font>

---



```python
profile_2 = ProfileReport(data_2, minimal=True)
profile_2.to_file(os.path.join(path, "report_2.html"))
```

```python
'''Data Year2'''
# Attr21 has 3164 (31.1%) missing values
# Attr37 has 4518 (44.4%) missing values
# Attr6 has 4455 (43.8%) zeros
# Attr59 has 4518 (44.4%) zeros
# Attr22 has 966 (9.5%) zeros
# Attr42 has 962 (9.5%) zeros
data_2 = data_2.drop(columns= ['Attr21','Attr37', 'Attr6', 'Attr59', 'Attr22','Attr42'], axis=1)

# Most of the features are skewed hence median is a better imputation strategy
data_2 = data_2.fillna(data_2.median()) 
```
